```python
import tweepy
import config
from datetime import datetime,timezone
import pytz
import pandas as pd
# basic data science modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime, os
import itertools
import time
client = tweepy.Client(bearer_token=config.BEARER_TOKEN)
query = '夫婦別姓'
counts = client.get_all_tweets_count(query=query,  start_time='2022-08-01T00:00:00Z',
        end_time='2022-08-31T00:00:00Z', granularity='day')
for count in counts.data:
    print(count)
```

```python
created_at = []
tweet_count = []

for count in counts.data:
    s = count['start']
    target = 'T'
    idx = s.find(target)
    idx = s.find(target)
    r = s[:idx]
    t = r.split('-')[2]
    count['created_at_day'] = t
    created_at.append(count['created_at_day'])
    tweet_count.append(count['tweet_count'])
```

```python
# create dataframe
cols = ['created_at','tweet_count']
tweets_counts_df = pd.DataFrame(list(zip(created_at,tweet_count)), columns = cols)
tweets_counts_df
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data = tweets_counts_df, x = 'created_at', y = 'tweet_count')

#データを月にする。
```
