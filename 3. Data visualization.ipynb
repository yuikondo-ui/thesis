```python
#全体ツイート

tweets_df = pd.read_csv("tweets_1.csv").drop_duplicates(subset="tweet_id")
tweets_df['created_at'] = pd.to_datetime(tweets_df['created_at'])
tweets_df['tweet_count'] = 1
tweets_df['created_at_day'] = tweets_df['created_at'].dt.strftime('%m-%d')

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data = tweets_df.groupby('created_at_day').sum().reset_index(), x = 'created_at_day', y = 'tweet_count')
```

```python
sns.lineplot(data = tweets_df.groupby('created_at_day').sum().reset_index(), x = 'created_at_day', y = 'tweet_count')

sns.lineplot(data = tweets_df[tweets_df["referenced_tweet_type"] == 'retweeted'].groupby('created_at_day').sum().reset_index(), x = 'created_at_day', y = 'tweet_count')


sns.catplot(y="referenced_tweet_type", data=tweets_df,
            kind="count",
            )

# Show plot
plt.show()
```
